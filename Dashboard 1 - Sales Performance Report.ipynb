# Dashboard 1

image.png

image.png


This dashboard presents a summary of DQStore's sales performance for the year 2017. Various metrics are displayed to provide a comprehensive overview of the business performance, including:

- `Total Sales`: Displays the total value of sales recorded throughout the year 2017.
- `Total Profit`: Shows the total profit earned from sales after deducting costs.
- `Profit Margin`: Calculates the percentage of profit compared to total sales, providing insights into the company's profitability efficiency.
- `Total Orders`: The total number of transactions or orders processed by DQStore during 2017.
- `Total Customers`: Represents the total number of customers who made purchases during the same period.
- `Average Sales Process Time`: Shows the average time taken from the start to the end of the sales process, giving an insight into operational efficiency.
- `Top 5 Customers`: The top five customers based on their total purchase value in 2017.
- `Sales by Location`: Displays sales data distributed by geographic location, helping map out markets and identify the best-performing areas.
- `Sales by Segment`: Sales analysis based on defined market segments, providing insights into how different segments contribute to overall sales.

With this detailed information, this dashboard aims to offer a clear understanding of DQStore's sales performance in 2017 and serves as a useful tool for strategic decision-making.
